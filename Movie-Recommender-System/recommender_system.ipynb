```python
# genres
# id
# keywords
# title
# overview
# cast
# crew

# we'll use these columns in our project, most useful for us

movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.head()
```

```python
# Now, we'll set only 3 columns out of 7 i.e. movie id, title and tags 
# for tags, we'll combine 5 columns i.e. overview, genres, keywords, cast, crew
```

```python
# removing null data

movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
# removing duplicate data

movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# convert this format to ['Action', 'Adventure'. 'Fantasy', 'Science Fiction']

# ast.literal_eval -> helps to convert list of strings to list only

import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# movies['cast'][0] in this we can see we have all information, but we don't want all this, we need only few things so like only 1st 3 actor names-

def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
# movies['crew'][0] same with this we only want director name-

def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# now we have to perform transformation in all of our columns values-
# eg- convert Sam Mendes into SamMendes like we have to remove spaces between names beacuse if we don't do that it will create 2 entitites for Sam Medes
# one for sam and one for mendes and for sure there will we another sam in list
# so if we search sam it will confuse between 2 different names as there will2 2 different etities/tags of sam, that's why we have to remove spaces
```

```python
movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# Now. we'll create new tags column and add 5 of these columns in this, like we discussed before-

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```
